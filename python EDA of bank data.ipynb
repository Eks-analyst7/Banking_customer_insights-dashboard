```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns 
import numpy as np

df = pd.read_csv('Banking.csv')
df
```

```python
import matplotlib.pyplot as plt
import seaborn as sns

for i, predictor in enumerate(['Nationality', 'Occupation', 'Fee Structure', 'Loyalty Classification',
                               'Estimated Income', 'Properties Owned', 'Risk Weighting',
                               'BRId', 'GenderId', 'IAId']):
    plt.figure(i)
    sns.countplot(data=df, x=predictor)
    plt.title(f'Count Plot of {predictor}')
    plt.xticks(rotation=45)
    plt.tight_layout()
    plt.show()
```

```python
numerical_cols = ['Estimated Income', 'Properties Owned', 'Risk Weighting']

for col in numerical_cols:
    plt.figure(figsize=(14, 4))

    
    plt.subplot(1, 2, 1)
    sns.histplot(df[col], kde=True)
    plt.title(f'Distribution of {col}')


    plt.subplot(1, 2, 2)
    sns.boxplot(x=df[col])
    plt.title(f'Boxplot of {col}')

    plt.tight_layout()
    plt.show()
```

```python
plt.figure(figsize=(10, 5))
sns.boxplot(data=df, x='Occupation', y='Estimated Income')
plt.xticks(rotation=45)
plt.title('Estimated Income across Occupations')
plt.tight_layout()
plt.show()
```

```python
plt.figure(figsize=(8, 6))
sns.heatmap(df[numerical_cols].corr(), annot=True, cmap='coolwarm')
plt.title('Correlation Heatmap')
plt.show()
```

```python
grouped_income = df.groupby('Occupation')['Estimated Income'].mean().sort_values(ascending=False)
print(grouped_income)
```

```python
grouped_risk = df.groupby('Loyalty Classification')['Risk Weighting'].mean()
print(grouped_risk)
```

```python
import matplotlib.pyplot as plt
import seaborn as sns

numerical_cols = ['Estimated Income', 'Properties Owned', 'Risk Weighting']

for col in numerical_cols:
    fig, axes = plt.subplots(1, 2, figsize=(12, 5))  # wider layout


    sns.histplot(df[col], kde=True, ax=axes[0], color='skyblue')
    axes[0].set_title(f'Distribution of {col}')
    axes[0].set_xlabel(col)


    sns.boxplot(x=df[col], ax=axes[1], color='salmon')
    axes[1].set_title(f'Boxplot of {col}')
    axes[1].set_xlabel(col)

    plt.tight_layout(pad=3.0)
    plt.show()
```
